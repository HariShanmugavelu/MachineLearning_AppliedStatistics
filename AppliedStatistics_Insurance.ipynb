```python
#4.a)Do charges of people who smoke differ significantly from the people who don't?
Charges_Smoker_Array=np.array(insurance[['charges','smoker']])
group1=Charges_Smoker_Array[:,1]=='yes'
group1=Charges_Smoker_Array[group1][:,0]

group2=Charges_Smoker_Array[:,1]=='no'
group2=Charges_Smoker_Array[group2][:,0]


#H0: x̄1 = x̄2, or x̄2 - x̄1 = 0, that is , there is no difference between the sample means
#HA: x̄2 < x̄1, or x̄2 - x̄1 < 0
#α = 0.05

#Method1- Considering the variance across the samples are same
t_statistic, p_value=ttest_ind(group1, group2)
print(t_statistic, p_value)

#Method2 - considering the variance across the samples are not same
u, p_value=mannwhitneyu(group1, group2)
print(u, p_value)

#Method3 - considering the variance across the samples are same but with detailed metrics
N1 = len(group1)
N2 = len(group2)
print(N1)
print(N2)
d1 = N1-1
d2 = N2-1
df = d1+d2
s1 = np.std(group1,ddof=1)
s2 = np.std(group2,ddof=1)
group1_bar = np.mean(group1)
group2_bar = np.mean(group2)

sp = np.sqrt((d1*s1**2 + d2*s2**2)/df)
se = sp*np.sqrt(1/N1 + 1/N2)
t = (group2_bar - group1_bar)/(se)
print("t-statistic", t)
t, p_twosided = stats.ttest_ind(group2, group1, equal_var=True)
print("t = ",t, ", p_twosided = ", p_twosided, ", p_onesided =", p_twosided/2)

#Ans: Since the p_value is less than 0.05, we reject the null hypothesis and conclude that there is significant difference 
#on charges of people who smoke when compared to those who don't
```

```python
#4.b) Does bmi of males differ significantly from that of females?
Charges_Smoker_Array=np.array(insurance[['bmi','sex']])
group1=Charges_Smoker_Array[:,1]=='male'
group1=Charges_Smoker_Array[group1][:,0]

group2=Charges_Smoker_Array[:,1]=='female'
group2=Charges_Smoker_Array[group2][:,0]


#H0: x̄1 = x̄2, or x̄2 - x̄1 = 0, that is , there is no difference between the sample means
#HA: x̄2 < x̄1, or x̄2 - x̄1 < 0
#α = 0.05

#Method1- Considering the variance across the samples are same
t_statistic, p_value=ttest_ind(group1, group2)
print(t_statistic, p_value)

#Method2 - considering the variance across the samples are not same
u, p_value=mannwhitneyu(group1, group2)
print(u, p_value)

#Method3 - considering the variance across the samples are same but with detailed metrics
N1 = len(group1)
N2 = len(group2)
print(N1)
print(N2)
d1 = N1-1
d2 = N2-1
df = d1+d2
s1 = np.std(group1,ddof=1)
s2 = np.std(group2,ddof=1)
group1_bar = np.mean(group1)
group2_bar = np.mean(group2)

sp = np.sqrt((d1*s1**2 + d2*s2**2)/df)
se = sp*np.sqrt(1/N1 + 1/N2)
t = (group2_bar - group1_bar)/(se)
print("t-statistic", t)
t, p_twosided = stats.ttest_ind(group2, group1, equal_var=True)
print("t = ",t, ", p_twosided = ", p_twosided, ", p_onesided =", p_twosided/2)

#Ans: Since the p_value is greater than 0.05, we accept null hypothesis and conclude that there is no significant difference
#in bmi of males when compared to bmi of females
```

```python
#4.c) Is the proportion of smokers significantly different in different genders?

#H0: The proportions are same
#HA: The proportions are significiantly different

cont = pd.crosstab(insurance["sex"], insurance["smoker"])
print(cont)
chi2_contingency(cont)

#Ans: From the chi-square statistic and pvalue, we can reject null hypothesis and claim that
#the proportions are significantly different
```

```python
#4.d) Is the distribution of bmi across women with no children, one child and two children, the same ?

#Null Hypothesis: Distribution of bmi across the samples are not same
#Alternate Hypothesis: Distribution of bmi across the samples are same

Bmi_Women_Children = np.array(insurance[['bmi', 'sex', 'children']])
group1=Bmi_Women_Children[:,1]=='female'
group1=Bmi_Women_Children[group1][:,[0,2]]
group1_Bmi_Child=group1[:,1]==0
group1_Bmi_Child=group1[group1_Bmi_Child][:,0]
print(np.mean(group1_Bmi_Child))
print(np.std(group1_Bmi_Child))


group2=Bmi_Women_Children[:,1]=='female'
group2=Bmi_Women_Children[group2][:,[0,2]]
group2_Bmi_Child=group2[:,1]==1
group2_Bmi_Child=group2[group2_Bmi_Child][:,0]
print(np.mean(group2_Bmi_Child))
print(np.std(group2_Bmi_Child))


group3=Bmi_Women_Children[:,1]=='female'
group3=Bmi_Women_Children[group3][:,[0,2]]
group3_Bmi_Child=group3[:,1]==2
group3_Bmi_Child=group3[group3_Bmi_Child][:,0]
print(np.mean(group3_Bmi_Child))
print(np.std(group3_Bmi_Child))

anova=stats.f_oneway(group1_Bmi_Child, group2_Bmi_Child, group3_Bmi_Child)
print(anova)

#Ans1: since the p-value across the samples is greater than 0.05, we accept the null hypothesis and conclude that 
#the distribution is not same 

#Ans2: No, the distribution of bmi across women with no children, one child and two children is not the same
#since the mean and standard deviation is not the same across the samples/groups
```
